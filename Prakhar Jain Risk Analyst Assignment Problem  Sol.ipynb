### Q.1 For all unique combinations of SYMBOL + EXPIRY_DT, find the cumulative value (VAL_INLAKH) traded in options (i.e. OPTION_TYPE is either CE or PE (and not XX)). Sort all such unique combinations in terms of cumulative option value traded.

```python
import pandas as pd
# Importing csv File In jupyter note Book
df = pd.read_csv('fo01AUG2017bhav.csv')
# Dropping Unnecessary Columns
df = df.drop('Unnamed: 15', axis=1)
# Dropping unnecessary Option type
df1 = df.drop(df.loc[df['OPTION_TYP'] == 'XX',['OPTION_TYP']],axis = 1 )
# grouping the SYMBOL and EXPIRY_DATE
df1 = df1.groupby(['SYMBOL','EXPIRY_DT'])['VAL_INLAKH'].agg({'Cumulative_VAL_INLAKH' : 'sum'})
df1.sort_values(['Cumulative_VAL_INLAKH'], ascending=False)

```

### Q.2  For all unique combinations of SYMBOL + EXPIRY_DT, report the (i) strike and (ii) value traded for the 5 nearest OTM call options of each group (i.e. same SYMBOL+EXPIRY_DT).      5 nearest OTM calls = call options (i.e. OPTION_TYPE=CE), and whose strike (STRIKE_PR) is higher than but nearest to SETTLE_PR of the underlying (i.e. SETTLE_PR of the instrument whose OPTION_TYPE = XX for the same SYMBOL + EXPIRY_DT)


### OTM calls


```python
df2 = df.set_index(['SYMBOL','EXPIRY_DT']).sort_values(['SYMBOL'], ascending=True)
# Set the condition For Option type 'XX'
df3 = df2.loc[df2['OPTION_TYP'] == 'XX',['SETTLE_PR']]  
# Set the condition For Option type 'CE'
df4 = df2.loc[df2['OPTION_TYP'] == 'CE',['OPTION_TYP', 'STRIKE_PR', 'VAL_INLAKH']]
for idx, df_select in df3.groupby(level=[0, 1]):
    if df4.index.isin([idx]).any():
        a = df3.loc[idx]['SETTLE_PR']
        b = df4.loc[idx]['STRIKE_PR']-a                # Finding STRIKE_PR  higher than but nearest to SETTLE_PR 
        b = df4.loc[idx][['STRIKE_PR', 'VAL_INLAKH']][b>0]
        
        print(b.nsmallest(5,'STRIKE_PR') )             #For combinations of SYMBOL + EXPIRY_DT,(i) strike and (ii) value traded  
                                                       #for the 5 nearest OTM call options of each group
        print('------------------------------------------------')
    else:
        continue
```

### Put

```python
# Following Similar procedure to calculate Put call
for idx, df_select in df3.groupby(level=[0, 1]):
    if df4.index.isin([idx]).any():
        a = df3.loc[idx]['SETTLE_PR']
        b = df4.loc[idx]['STRIKE_PR']-a
        b = df4.loc[idx][['STRIKE_PR', 'VAL_INLAKH']][b<0]
        # Get combinations of SYMBOL + EXPIRY_DT,(i) strike and (ii) value traded  
        #for the 5 nearest Put call options of each group
        print(b.nsmallest(5,'STRIKE_PR'))                       
        print('------------------------------------------------')
    else:
        continue
```

### Q4. Download the bhavcopy for consecutive dates from the below links. Report the SYMBOL+EXPIRY_DT groups with highest increase in trading volume relative to the previous date.

```python
# The Dataset used in this calculation is 'fo01AUG2017bhav.csv' and 'fo02AUG2017bhav.csv'

import pandas as pd
df5 = pd.read_csv('fo02AUG2017bhav.csv')
df5 = df5.drop('Unnamed: 15', axis=1)

df5 = df5[['SYMBOL', 'EXPIRY_DT', 'CONTRACTS']]
df = pd.read_csv('fo01AUG2017bhav.csv')
# Dropping Unnecessary Columns
df = df.drop('Unnamed: 15', axis=1)
df = df[['SYMBOL', 'EXPIRY_DT', 'CONTRACTS']]
# Merging two DATA Frame 
df6 = pd.merge_asof(df, df5,on = 'SYMBOL',left_index=True, right_index=True)

# Finding Difference between Volume trading (CONTRACTS) of two DATA Frames

df6['Diff in trading volume'] = df6['CONTRACTS_y']-df6['CONTRACTS_x']
df6 = df6[['SYMBOL','EXPIRY_DT_x','CONTRACTS_x','CONTRACTS_y','Diff in trading volume']]

#renaming the Column EXPIRY_DT_x as EXPIRY_DT
df6 = df6.rename(columns = {'EXPIRY_DT_x':'EXPIRY_DT'})

# Setting the Index SYMBOL and EXPIRY_DT
df6 = df6.set_index(['SYMBOL', 'EXPIRY_DT'])

# Sorting the Dataset in Decreasing order according to the Diiference in Trading Volume
df6.sort_values(['Diff in trading volume'], ascending=False)
```
